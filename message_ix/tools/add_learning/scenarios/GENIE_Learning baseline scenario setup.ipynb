```python
import ixmp
import message_ix
import numpy as np
import pandas as pd
import yaml

from collections.abc import Mapping
from itertools import repeat
from message_ix.models import MESSAGE_ITEMS
from message_ix.utils import make_df

from message_ix.tools.add_learning import add_learning


%matplotlib inline
```

```python
mp = ixmp.Platform("local")
```

## Making a clone of the existing scenario 'baseline'

Dump scenario to excel nad create a scenario from excel

```python
base_is_new = "yes_sandbox" # select "Yes" or "No"

if base_is_new == "yes_engage":
    data_file = "ENGAGE_SSP2_v417_NPi2020-con-prim-dir-ncr.xlsx"

    # Creating a new, empty scenario
    base = message_ix.Scenario(
        mp, model="GENIE", scenario="baseline", version="new"
    )
    # Fill the empty scenario with data from excel
    base.read_excel(data_file, add_units=True, init_items=True)

elif base_is_new == "yes_sandbox":
    data_file = "GENIE_sandbox_baseline.xlsx"

    # Creating a new, empty scenario
    base = message_ix.Scenario(
        mp, model="GENIE_sandbox", scenario="baseline", version="new"
    )
    
    # Dump the scenario data to excel
    base.read_excel(data_file, add_units=True, init_items=True)

elif base_is_new == "no_engage":
    base = message_ix.Scenario(
        mp, model="GENIE", scenario="baseline",
    )

else:
    base = message_ix.Scenario(
        mp, model="GENIE_sandbox", scenario="baseline",
    )
```

Clone baseline scenario and add learning parameters

```python
scen = base.clone(
    "GENIE_sandbox",
    "learning",
    "GENIE baseline scenario with learning",
    keep_solution=False,
)

scen.check_out()
```

```python
scen.par('inv_cost', filters={'technology':['dac_1_exports','dac_2_exports','dac_3_exports','dac_4_exports','dac_5_exports'],
                              'node_loc':'R11_WEU','year_vtg':2025})

scen.par('relation_activity', filters={'technology':'dac_1_exports'})
```

## Add learning parameters to the scenario

To add learning parameters to the scenario, one needs to include `add_learning(Scenario)` to the script, with `Scenario` is the name of scenario object that you want to run. In this example, we want to run `scen`, thus:

```python
add_learning(scen)
```

## Time to Solve the Model

```python
scen.commit(comment="GENIE_sandbox baseline with learning")
scen.set_as_default()
```

```python
scen.solve(gams_args =["--learningmode=0"]) #
```
